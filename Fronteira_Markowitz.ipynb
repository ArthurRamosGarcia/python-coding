```python
import pandas as pd
import numpy as np
```

```python
composicao_ibov = pd.read_excel('IBOVDia_16-06-25.xlsx')
```

```python
composicao_ibov.columns=['Código', 'Empresa', 'Valor', 'Peso (%)']
```

```python
composicao_ibov=composicao_ibov.iloc[:-2]
composicao_ibov=composicao_ibov.iloc[1:]
```

```python
ativos=composicao_ibov['Código'].values
```

```python
ativos=[ativo +".SA" for ativo in ativos]
```

```python
ativos
```

```python
import yfinance as yf
```

```python
fechamentos=yf.download(ativos, start='2022-01-01', end='2022-12-31')
```

```python
fechamentos.dropna(axis=True, inplace=True)
```

```python
fechamentos.head()
```

```python
retornos=fechamentos['Close'].pct_change()
```

```python
retornos.head()
```

```python
matriz_cov=retornos.cov()
```

```python
matriz_cov.head()
```

```python
[tam, qtd_ativos] = retornos.shape
```

```python
qtd_pontos_fronteira = 25
```

```python
pontos_front=np.linspace(0,1,qtd_pontos_fronteira)
```

```python
pontos_front
```

```python
media_retornos=retornos.mean().values
```

```python
def is_pos_eigenvectors(a):
  M=np.matrix(a)
  return np.all(np.linalg.eigvals(M)>0)
```

```python
is_pos_eigenvectors(matriz_cov)
```

```python
from scipy.optimize import Bounds
```

```python
bounds=Bounds(np.zeros(qtd_ativos), np.ones(qtd_ativos))
```

```python
bounds
```

```python
def h(x):
  return sum(x)-1
```

```python
cons = ({'type': 'eq', 'fun': h})
```

```python
x0 = (1/qtd_ativos)*np.ones(qtd_ativos)
```

```python
x0
```

```python
def calcula_retorno(x):
  return(media_retornos.dot(x))
```

```python
def calcula_risco(x):
  return (x.dot(matriz_cov).dot(x.T))
```

```python
from scipy.optimize import minimize
```

```python
def otimizacao(aux_ponto_front):
  def funcao_objetivo(x, ponto_front=aux_ponto_front):
    retorno=calcula_retorno(x)
    risco=calcula_risco(x)
    return ponto_front*risco-(1-ponto_front)*retorno
  xmin=minimize(funcao_objetivo, x0, bounds=bounds, constraints=cons)
  return xmin.x
```

```python
def constroi_fronteira():
  carteiras_front=[]
  retorno=np.zeros(qtd_pontos_fronteira)
  risco=np.zeros(qtd_pontos_fronteira)
  pos = 0
  for ponto_front in pontos_front:
    carteira=otimizacao(ponto_front)
    carteiras_front.append(carteira)
    retorno[pos]=calcula_retorno(carteira)
    risco[pos]=calcula_risco(carteira)
    pos += 1
  return retorno, risco, carteiras_front
```

```python
retorno, risco, carteiras_front = constroi_fronteira()
```

```python
from bcb import sgs
```

```python
selic = sgs.get({'selic':432, 'igp-m':189}, start='2022-01-01')
```

```python
selic['selic_diaria'] = (1+selic['selic']/100)**(1/252)-1
```

```python
def calcula_indice_sharpe(x):
  retorno=calcula_retorno(x)
  risco=calcula_risco(x)
  return -1*((retorno-(selic['selic_diaria']).mean())/risco)
```

```python
min_sharpe=minimize(calcula_indice_sharpe, x0, method='SLSQP', bounds=bounds, constraints=cons)
```

```python
carteira_sharpe_otimo=min_sharpe.x
```

```python
risco_sharpe=calcula_risco(carteira_sharpe_otimo)
```

```python
retorno_sharpe=calcula_retorno(carteira_sharpe_otimo)
```

```python
print("Índice de Sharpe: ",round(((retorno_sharpe-(selic['selic_diaria']).mean())/risco_sharpe),2))
```
