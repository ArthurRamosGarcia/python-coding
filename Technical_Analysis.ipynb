```python
import yfinance as yf
import plotly.graph_objects as go
```

```python
amzn=yf.download("AMZN", start='2020-01-01')
amzn.columns=amzn.columns.get_level_values(0)
amzn.head()
```

```python
fig=go.Figure()
fig.add_trace(go.Scatter(x=amzn.index, y=amzn['Close'], name='Fechamento'))
fig.update_layout(template='simple_white', width=900, height=300)
fig.show()
```

```python
!pip install ta
```

```python
from ta.momentum import RSIIndicator
```

```python
rsi = RSIIndicator(close=amzn['Close'], window=14)

amzn['rsi'] = rsi.rsi()
amzn.dropna(inplace=True)
amzn.head()
```

```python
from plotly.subplots import make_subplots
fig = make_subplots(rows=2, cols=1, shared_xaxes=True, row_width=[0.2, 0.7])
fig.add_trace(go.Scatter(x=amzn.index, y=amzn['Close'], name='Fechamento'), row=1, col=1)
fig.add_trace(go.Scatter(x=amzn.index, y=amzn['rsi'], name='RSI'), row=2, col=1)
fig.update_layout(template='plotly_dark', width=1300, height=500)
fig.show()
```

```python
from ta.momentum import StochasticOscillator
```

```python
so=StochasticOscillator(high=amzn['High'], low=amzn['Low'], close=amzn['Close'], window=14)
amzn['so']=so.stoch()
amzn.dropna(inplace=True)
amzn.head()
```

```python
fig=make_subplots(rows=4, cols=1, row_width=[0.4, 0.4, 0.4, 0.4])
fig.add_trace(go.Scatter(x=amzn.index, y=amzn['Close'], name='Fechamento'), row=1, col=1)
fig.add_trace(go.Scatter(x=amzn.index, y=amzn['rsi'], name='RSI'), row=2, col=1)
fig.add_trace(go.Scatter(x=amzn.index, y=amzn['so'], name='Oscilador Estocástico'), row=3, col=1)
fig.add_trace(go.Scatter(x=amzn.index, y=amzn['so'].rolling(window=21).mean(), name='SO_MM21'), row=3, col=1)
fig.update_layout(template='plotly_dark', width=1300, height=500)
fig.show()
```

```python
from ta.momentum import ROCIndicator
```

```python
roc=ROCIndicator(close=amzn['Close'], window=21)
amzn['roc']=roc.roc()
amzn.dropna(inplace=True)
amzn.head()
```

```python
fig=make_subplots(rows=4, cols=1, row_width=[0.4, 0.4, 0.4, 0.4])
fig.add_trace(go.Scatter(x=amzn.index, y=amzn['Close'], name='Fechamento'), row=1, col=1)
fig.add_trace(go.Scatter(x=amzn.index, y=amzn['rsi'], name='RSI'), row=2, col=1)
fig.add_trace(go.Scatter(x=amzn.index, y=amzn['so'], name='Oscilador Estocástico'), row=3, col=1)
fig.add_trace(go.Scatter(x=amzn.index, y=amzn['so'].rolling(window=21).mean(), name='SO_MM21'), row=3, col=1)
fig.add_trace(go.Scatter(x=amzn.index, y=amzn['roc'].rolling(window=21).mean(), name='change_21'), row=4, col=1)
fig.update_layout(template='plotly_dark', width=1300, height=500)
fig.show()
```

```python
from ta.momentum import PercentagePriceOscillator
```

```python
ppo = PercentagePriceOscillator(close=amzn['Close'], window_slow=15, window_fast=5, window_sign=9)
amzn['ppo'] = ppo.ppo()
amzn['ppo_hist'] = ppo.ppo_hist()
amzn['ppo_line'] = ppo.ppo_signal()
amzn.dropna(inplace=True)
amzn.head()
```

```python
fig=make_subplots(rows=6, cols=1, row_width=[0.4, 0.4, 0.4, 0.4, 0.4, 0.4])
fig.add_trace(go.Scatter(x=amzn.index, y=amzn['Close'], name='Fechamento'), row=1, col=1)
fig.add_trace(go.Scatter(x=amzn.index, y=amzn['rsi'], name='Relative Strengh Index'), row=2, col=1)
fig.add_trace(go.Scatter(x=amzn.index, y=amzn['so'], name='Oscilador Estocástico'), row=3, col=1)
fig.add_trace(go.Scatter(x=amzn.index, y=amzn['so'].rolling(window=21).mean(), name='SO_MM21'), row=3, col=1)
fig.add_trace(go.Scatter(x=amzn.index, y=amzn['roc'].rolling(window=21).mean(), name='Varicao_21d'), row=4, col=1)
fig.add_trace(go.Scatter(x=amzn.index, y=amzn['ppo'].rolling(window=21).mean(), name='Percentage Price Oscilator'), row=5, col=1)
fig.add_trace(go.Scatter(x=amzn.index, y=amzn['ppo_line'].rolling(window=21).mean(), name='ppo_signal'), row=5, col=1)
fig.add_trace(go.Scatter(x=amzn.index, y=amzn['ppo_hist'].rolling(window=21).mean(), name='Rompimento Sinal'), row=6, col=1)
fig.update_layout(template='plotly_dark', width=1300, height=500)
fig.show()
```

```python
from ta.volume import VolumeWeightedAveragePrice
```

```python
vwap=VolumeWeightedAveragePrice(high=amzn['High'], low=amzn['Low'], close=amzn['Close'], volume=amzn['Volume'])
amzn['vwap']=vwap.volume_weighted_average_price()
amzn.dropna(inplace=True)
amzn.head()
```

```python
fig=make_subplots(rows=7, cols=1, row_width=[0.4, 0.4, 0.4, 0.4, 0.4, 0.4, 0.4])
fig.add_trace(go.Scatter(x=amzn.index, y=amzn['Close'], name='Fechamento'), row=1, col=1)
fig.add_trace(go.Scatter(x=amzn.index, y=amzn['rsi'], name='Relative Strengh Index'), row=2, col=1)
fig.add_trace(go.Scatter(x=amzn.index, y=amzn['so'], name='Oscilador Estocástico'), row=3, col=1)
fig.add_trace(go.Scatter(x=amzn.index, y=amzn['so'].rolling(window=21).mean(), name='SO_MM21'), row=3, col=1)
fig.add_trace(go.Scatter(x=amzn.index, y=amzn['roc'].rolling(window=21).mean(), name='Varicao_21d'), row=4, col=1)
fig.add_trace(go.Scatter(x=amzn.index, y=amzn['ppo'].rolling(window=21).mean(), name='Percentage Price Oscilator'), row=5, col=1)
fig.add_trace(go.Scatter(x=amzn.index, y=amzn['ppo_line'].rolling(window=21).mean(), name='ppo_signal'), row=5, col=1)
fig.add_trace(go.Scatter(x=amzn.index, y=amzn['ppo_hist'].rolling(window=21).mean(), name='Rompimento Sinal'), row=6, col=1)
fig.add_trace(go.Scatter(x=amzn.index, y=amzn['vwap'].rolling(window=21).mean(), name='Preço Médio Ponderado (VWAP)'), row=7, col=1)
fig.update_layout(template='plotly_dark', width=1300, height=500)
fig.show()
```
