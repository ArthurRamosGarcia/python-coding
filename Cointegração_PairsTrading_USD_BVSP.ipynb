```python
ativo1='USDBRL=X'
ativo2='^BVSP'

tickers=[ativo1, ativo2]
ativos=yf.download(tickers, start='2010-07-01', end='2025-06-30')['Close']
ativos.dropna(inplace=True)
ativos.head()
```

```python
fig=go.Figure()
fig.add_trace(go.Scatter(name="Dlar", x=ativos.index, y=ativos[ativo1]/ativos[ativo1].iloc[0]))
fig.add_trace(go.Scatter(name="ndice Bovespa", x=ativos.index, y=ativos[ativo2]/ativos[ativo2].iloc[0]))
fig.update_layout(template='simple_white', height=300, width=900)
fig.show()
```

```python
fig=go.Figure()
fig.add_trace(go.Scatter(x=ativos[ativo1], y=ativos[ativo2], mode='markers'))
fig.update_layout(template='simple_white', height=300, width=900)
fig.show()
```

```python
fig=go.Figure()
fig.add_trace(go.Scatter(x=ativos[ativo1].pct_change(), y=ativos[ativo2].pct_change(), mode='markers'))
fig.update_layout(template='simple_white', height=300, width=900)
fig.show()
```

```python
#Teste de Estacionariedade dos Resduos da Regresso
score, pvalue, _ = coint(ativos[ativo1], ativos[ativo2])
print('Score: %f' % score)
print('p-value: %f' % pvalue)
```

```python
x1=ativos[ativo1]
x2=ativos[ativo2]
x1=sm.add_constant(x1)
resultado=sm.OLS(x2, x1).fit()
x1=x1[ativo1]
print(resultado.summary())
```

```python
beta=resultado.params[1]
spread=x2-beta*x1
fig=go.Figure()
fig.add_trace(go.Scatter(x=ativos.index, y=spread))
fig.update_layout(template='simple_white', height=300, width=900)
fig.show()
```

```python
teste_residuos=adfuller(spread)
print('ADF Statistic: %f' % teste_residuos[0])
print('p-value: %f' % teste_residuos[1])
```

```python
z_score=(spread-spread.mean())/np.std(spread)
```

```python
fig=go.Figure()
fig.add_trace(go.Scatter(x=ativos.index, y=z_score))
fig.update_layout(template='simple_white', height=300, width=900)
fig.show()
```

```python
#Oscilador de Resduos em mltiplos de Desvio-Padro
from plotly.subplots import make_subplots
fig=make_subplots(rows=2, cols=1)
fig.add_trace(go.Scatter(x=ativos.index, y=ativos[ativo1] / ativos[ativo1].iloc[0], name="Taxa de Cmbio (R$/US$)"), row=1, col=1)
fig.add_trace(go.Scatter(x=ativos.index, y=ativos[ativo2]/ativos[ativo2].iloc[0], name="ndice Bovespa"), row=1, col=1)
fig.add_trace(go.Scatter(x=z_score.index, y=z_score, name="Z-Score"), row=2, col=1)
fig.update_layout(template='simple_white', height=500, width=900)
fig.show()
```

```python
import vectorbt as vbt
```

```python
# Parmetros

Caixa = 100000
PctOrdem1=0.20
PctOrdem2=0.50
Banda_Sup=1.00 #Sinal para Venda (Em n desvios-padro)
Banda_Inf=-1.00 #Sinal para Compra (Em n desvios-padro)
```

```python
sinal_short=(z_score>Banda_Sup).rename('sinal_short')
sinal_long=(z_score<Banda_Inf).rename('sinal_long')
```

```python
vbt_sinal_short, vbt_sinal_long = pd.Series.vbt.signals.clean(sinal_short, sinal_long, entry_first=False, broadcast_kwargs=dict(columns_from='keep'))
```

```python
tickers_coluna=pd.Index([ativo1], name='tickers')
vbt_ordem=pd.DataFrame(index=ativos.index, columns=tickers_coluna)
vbt_ordem[ativo1]=np.nan
vbt_ordem[ativo2]=np.nan
vbt_ordem.loc[vbt_sinal_short, ativo1]=+PctOrdem1
vbt_ordem.loc[vbt_sinal_long, ativo1]=-PctOrdem1
vbt_ordem.loc[vbt_sinal_short, ativo2]=-PctOrdem2
vbt_ordem.loc[vbt_sinal_long, ativo2]=+PctOrdem2

vbt_ordem=vbt_ordem.vbt.fshift(1)
print(vbt_ordem[~vbt_ordem.isnull().any(axis=1)])
```

```python
def portfolio_pairs_trading():
  return vbt.Portfolio.from_orders(
      ativos,
      size=vbt_ordem,
      price=ativos,
      size_type='targetpercent',
      val_price=ativos.vbt.fshift(1),
      init_cash=Caixa,
      fees=0.0002,
      group_by=True,
      call_seq='auto',
      freq='d')

vbt_pf=portfolio_pairs_trading()
```
